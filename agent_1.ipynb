Step 1: Define the AI Agent

```python
from crewai import Agent
from langchain_openai import ChatOpenAI
from google.colab import userdata

import os

# Retrive open ai key
os.environ['OPENAI_API_KEY'] = userdata.get('sg-test-1')

# Create an instance of OpenAI's LLM
openai_llm = ChatOpenAI(
    model_name="gpt-4o",
    temperature=0.7
)

#define the agent
chat_bot = Agent(role='Critical Thinker',
                 goal='Analze the text and identify if any conflicting information within',
                 llm=openai_llm,
                 verbose=False,
                 backstory=(
                     'You are a critical thinker who understands details very well.'
                     "You can identify conflicing statemes, information in given text"
                  )
                 )
```

Step 2: Define the Task

```python
from crewai import Task

# Define the task
chat_task = Task(
    description=('Find if there are any conflicting statement / information in text. \n Text : \n{text}'),
    expected_output="Respond with 'conflict' / 'no conflict'",
    agent=chat_bot,
)
```

Step 3: Initialize the Crew

```python
from crewai import Crew, Process
from google.colab import userdata
import os

# Initialize the Crew
crew = Crew(
    agents = [chat_bot],
    tasks=[chat_task],
)
```

Step 4: Run the Crew and Get the Response

```python
# kickoff the Crew with input query
Text = "After a long day at office, I was going back home in the late evening. While I was going home at 8 pm, I met my friend in the office"

result = crew.kickoff(inputs={'text': Text})

#print the response
print("Response:", result)
```
